```python
def Essy_calculate():
    print('YOU ARE WELCOME TO Essy Mini project!!!')
    print('PROGRAMMED BY Ikiseh EstherJoan')
    print('THE OPERATIONS ARE AS FOLLOWS')
    print("Main Menu")
    print("1. ADDITION +")
    print("2. SUBTRACTION - ")
    print("3. MULTIPLICATION *")
    print("4. DIVISION")
    print("5. AVERAGE")
    print("6. QUIT")
    print("NOTE THAT YOU ARE TO ENTER THE RESPECTIVE NUMBERS FOR EVERY OPERATIONS")

def sum_of_numbers():
    numbers = []
    while True:
        try:
            user_input = input("Enter a numeric number (or 'done' to finish): ")

            if user_input.lower() == 'done':
                total = sum(numbers)
                print("Sum of the numbers:", total)
                break

            number = float(user_input)
            numbers.append(number)

        except ValueError:
            print("Invalid input. Please enter a valid number.")

def subtract_numbers():
    try:
        base_number = float(input("Enter the base number: "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return

    result = base_number

    while True:
        try:
            user_input = input("Enter a number to subtract (or 'done' to finish): ")

            if user_input.lower() == 'done':
                break

            number_to_subtract = float(user_input)
            result -= number_to_subtract
            print(f'{result} - {number_to_subtract} ')

        except ValueError:
            print("Invalid input. Please enter a valid number.")

    print("Final result:", result)

def calculate_product():
    try:
        product = 1
        val = []

        while True:
            user_input = input("Enter a number (or 'done' to finish): ")

            if user_input.lower() == 'done':
                print('the multiplicated values are as follows:', val)
                print("Product of the numbers:", product)
                break

            number = float(user_input)
            val.append(number)
            product *= number

    except ValueError:
        print("Invalid input. Please enter valid numbers.")

def divide_numbers():
    try:
        base_number = float(input("Enter the base number: "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return

    result = base_number

    while True:
        try:
            user_input = input("Enter a number to divide (or 'done' to finish): ")

            if user_input.lower() == 'done':
                break

            number_to_divide = float(user_input)
            result /= number_to_divide
            print(f'{result} / {number_to_divide} ')

        except ValueError:
            print("Invalid input. Please enter a valid number.")
        except ZeroDivisionError:
            print("Error: Cannot divide by zero. Please enter a non-zero divisor.")


    print("Final result:", result)

def calculate_average():
    try:
        total = 0
        count = 0

        while True:
            user_input = input("Enter a number (or 'done' to finish): ")

            if user_input.lower() == 'done':
                break

            number = float(user_input)
            total += number
            count += 1

        if count > 0:
            average = total / count
            print("Average of the numbers:", average)
        else:
            print("No numbers entered.")

    except ValueError:
        print("Invalid input. Please enter valid numbers.")

def main():
    Essy_calculate()
    options = None
    while True:
        options = input("Enter your options: ")

        if options == '1':
            sum_of_numbers()

        elif options == '2':
            subtract_numbers()

        elif options == '3':
            calculate_product()

        elif options == '4':
            divide_numbers()

        elif options == '5':
            calculate_average()

        elif options == '6':
            print("THANKS PROGRAM QUITTED")
            print("GOOD BYE")
            break
        else:
            print("Invalid options. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
